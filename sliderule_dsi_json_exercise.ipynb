## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# assign json file to a variable
json_data = pd.read_json('data/world_bank_projects.json')

# look for keys used for diagnosis
json_data.keys()
```

```python
# create DataFrame with columns of only 'countryname' and 'project_name'
country_and_projects = pd.concat([json_data['countryname'], 
                                  json_data['project_name']], 
                                  axis = 1, 
                                  keys = ['countryname','project_name'])

# sort DataFrame by 'countryname' and reset the index
sorted_df = country_and_projects.sort_values('countryname')
sorted_df = sorted_df.reset_index()

# view DataFrame
sorted_df
```

```python
# Describe used to check number of non-null values == number of rows in sorted_df 
# Describe also proves all values are unique
sorted_df['project_name'].describe()
```

```python
# Count the number of each country value and print top 10 occuring results
unique_country_values = pd.value_counts(sorted_df['countryname'])
print(unique_country_values.head(10))
```

```python
# Load json file
data = json.load(open('data/world_bank_projects.json'))

# normalize json file to get dictionaries into DataFrame
theme = json_normalize(data, 'mjtheme_namecode')
theme
```

```python
# Empty dictionary to be filled with key=code and value=name
ref_dict = {}

# loop through only unique values of theme['code']
for code in theme['code'].unique():
    
    # Build DataFrame to hold each unique value of theme['code']
    subset_df = theme[theme['code'] == code]
    
    # Use unique code entries as key and add 'name' as value
    ref_dict[code] = subset_df['name'].value_counts().index[0]
print(ref_dict)
```

```python
# Organize number of times theme used into variable
theme_organized = theme['code'].value_counts()

# Set name of theme for top 10 themes as variable
top_ten_themes_ind = theme_organized[:10].index

# Loop through top ten themes and print out theme name with count of theme used
for i, count in enumerate(theme_organized[:10]):
    print(ref_dict[top_ten_themes_ind[i]], count)
```

```python
# Create DataFrame from ref_dict for filling empty theme values
join_dict = pd.DataFrame.from_dict(ref_dict, orient='index')
join_dict
```

```python
# Join join_dict DataFrame and theme DataFrame via the Code index
final_df = theme.join(join_dict, on='code')
final_df
```

```python
# Drop the column 'name' that holds empty string values
final_df = final_df.drop('name', axis=1)
final_df
```

```python
# Rename column of themes to 'theme'
final_df = final_df.rename(columns= {0: 'theme'})

final_df.head()
```
